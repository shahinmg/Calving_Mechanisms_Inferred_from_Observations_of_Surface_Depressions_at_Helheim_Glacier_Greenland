```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Mon Jul  8 14:25:06 2024

@author: laserglaciers
"""
import os
import pandas as pd
import geopandas as gpd
import numpy as np
from subProcessUtils import date_sort
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from pathlib import Path
from shapely.geometry import Point
import numpy as np
import string
import matplotlib as mpl
import matplotlib.pylab as pylab
from matplotlib.ticker import MultipleLocator
import matplotlib.ticker as ticker
```

```python
flexure_data_path = '../data/tides/flexure_vs_non-flexure_calving.csv'
flexure_data = pd.read_csv(flexure_data_path ,parse_dates=(['first_scan', 'last_scan']))

params = {'legend.fontsize': 'x-large',
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'xx-large',
         'ytick.labelsize':'xx-large'}
pylab.rcParams.update(params)



root_path = '/media/laserglaciers/upernavik/leigh_agu_2023/atlas_data/'
term_position_path = f'{root_path}atlas_terminus_position_row_30_cleaner.csv'
plot_dict = {'c':'tab:blue',
              'markersize':5,
              'markeredgecolor':'k',
              'markeredgewidth':0.3}

term_position_df = pd.read_csv(term_position_path,parse_dates=['date'])

term_data_path_leigh = '../data/atlas_its_live_comp/termDONE_raw.csv'
term_position_df = pd.read_csv(term_data_path_leigh,parse_dates=['Date'])
label_dict = {'fontsize': 12}
tick_size = 15

# Combined atlas, itslive on top and term position on botton
fig5, ax5 = plt.subplots(3,1, sharex=True, figsize=(16,8))
ax5 = ax5.flatten()



# ATLAS vel
atlas_vel_path = '../data/atlas_its_live_comp//ATLAS_abs_H_d8.pkl'
atlas_vel = pd.read_pickle(atlas_vel_path)
atlas_vel = atlas_vel.iloc[:,1:]
atlas_vel = atlas_vel.T
atlas_vel.index = pd.to_datetime(atlas_vel.index)
#remove bad data point; weird bulls eye artifact; likely noise from air
atlas_vel = atlas_vel[atlas_vel<50]
atlas_vel.rename({0:'v'}, axis=1, inplace=True)
ax5[0].plot(atlas_vel.index, atlas_vel.values, 'o',
           color='tab:blue', markeredgecolor='k', markeredgewidth = 0.3, label='ATLAS', linewidth=0.5, zorder=2)






ax5[0].grid(linestyle='--',zorder=1)
its_live_path = '/media/laserglaciers/upernavik/flexure_manuscipt/csv/helheim_velocity_its_live_review.pkl'
# its_live_df = pd.read_csv(its_live_path, parse_dates=['mid_date'])

its_live_df = pd.read_pickle(its_live_path)



its_live_df['v'] = its_live_df['v']/365
# its_live_df['v'] = its_live_df['v'][its_live_df['v']<50] # filter major outliers
its_live_df = its_live_df[its_live_df['v']>5] # filter out very low values
time_delta = pd.Timedelta(50,'D')
its_live_df['date_dt'] = its_live_df['date_dt'][its_live_df.date_dt < time_delta]

atlas_start = term_position_df.Date[0]
atlas_time_mask = (its_live_df.index > atlas_start) & (its_live_df.index < atlas_vel.index[-1])
its_live_df_atlas_time = its_live_df[atlas_time_mask]
its_live_df_atlas_time = its_live_df_atlas_time[(its_live_df_atlas_time['v'] < 50) & (its_live_df_atlas_time['v_error'] < 600)] # check filter


its_live_plot_dict = {'color':'tab:orange', 
                      'edgecolor': 'k',
                      'zorder': 2,
                      'linewidth': 0.3}

ax5[0].scatter(its_live_df_atlas_time.index, its_live_df_atlas_time['v'],label='ITS_LIVE', **its_live_plot_dict)



h, l = ax5[0].get_legend_handles_labels()
h, l = h[-2:], l[-2:]
ax5[0].legend(h, l,loc='upper center',ncol=2)

ax5[0].set_title(r'ATLAS and ITS_LIVE Velocity')
ax5[0].set_ylabel(r'Velocity (m d$^{-1}$)')

vel_xlims = ax5[0].get_xlim()


plot_dict_term = {'c':'tab:green',
              'markersize':5,
              'markeredgecolor':'k',
              'markeredgewidth':0.3}

ax5[1].plot(term_position_df.Date, term_position_df.terminus,'-o', **plot_dict_term)
km = r'km$^{2}$'
ax5[1].set_ylabel(f'{km}\nretreated      advanced')
ax5[1].set_ylim(-2.4, 2.4)

ax5[1].grid(linestyle='--',zorder=1)

# ax5[1].set_xlim(vel_xlims)
# ax5[1].sharex(ax5[0])
ax5[1].set_title(r'Terminus Position')

#%% bottom plot

its_live_rolling_dict = {'c':'tab:orange',
              'markersize': 5,
              'markeredgecolor':'k',
              'linewidth': 0.5,
              'zorder': 1}



atlas_rolling_dict = {'c':'tab:blue',
              'markersize': 5,
              'markeredgecolor':'k',
              'linewidth': 2,
              'zorder': 2}

window = pd.Timedelta(1,'W')
atlas_years = np.arange(2015,2024)
atlas_years_split = {}
for year in atlas_years:
    year_mask = atlas_vel.index.year == year
    atlas_year = atlas_vel[year_mask]
    atlas_year_rolling = atlas_year.rolling(window=window).mean()
    atlas_years_split[year] = atlas_year_rolling
    
    ax5[2].plot(atlas_year_rolling.index, atlas_year_rolling.values,'-',**atlas_rolling_dict)
    
its_live_rolling = its_live_df_atlas_time['v'].rolling(window=window).mean()
its_live_rolling_dict = {}
atlas_time_mask_rolling = (its_live_rolling.index > atlas_vel.index[0]) & (its_live_rolling.index < atlas_vel.index[-1])



for year in atlas_years:
    
    year_sample = atlas_years_split[year] 
    start_date = year_sample.index[0]
    end_date = year_sample.index[-1]
    year_mask = (its_live_rolling.index >= start_date) & (its_live_rolling.index <= end_date)
    its_live_rolling_year = its_live_rolling[year_mask]
    its_live_rolling_dict[year] = its_live_rolling_year
    
    # ax5[1].plot(its_live_rolling_year.index, its_live_rolling_year.values, '-', zorder=1, color='tab:orange',
    #             markersize=5, markeredgecolor='k', linewidth=2)

its_live_rolling_atlas_mask = its_live_rolling[atlas_time_mask_rolling]
ax5[2].plot(its_live_rolling.index, its_live_rolling.values, '-', zorder=1, color='tab:orange',
            markersize=5, markeredgecolor='k', linewidth=2)

ax5[2].grid(linestyle='--',zorder=1)
ax5[2].set_ylabel(r'Velocity (m d$^{-1}$)')

top_vel_ylims = ax5[0].get_ylim()
ax5[2].set_ylim(top_vel_ylims)
ax5[2].set_title(r'1-Week Rolling Average')


alphabet = list(string.ascii_lowercase)
text_dict = {'fontsize':15,
             'fontweight': 'bold'}

ml = MultipleLocator(2)
ax5[0].yaxis.set_minor_locator(ticker.AutoMinorLocator(2))


ax5[1].set_title(r'Terminus Position')

for i, axis in enumerate(ax5):
    axis.set_xlim(pd.to_datetime('2001'), pd.to_datetime('2024'))
    axis.tick_params(axis='both',labelsize = tick_size)
    # axis.tick_params(axis='x', which='minor', bottom=False)
    axis.xaxis.set_minor_locator(ticker.AutoMinorLocator(2))
    text_label = axis.text(.01, .99, alphabet[i], ha='left', va='top', transform=axis.transAxes, **text_dict)

plt.tight_layout()

```
